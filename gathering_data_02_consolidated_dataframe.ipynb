 Получившийся список должен содержать в себе минимум:

    *Наименование вакансии
    *Предлагаемую зарплату (отдельно мин. и отдельно макс.)
    *Ссылку на саму вакансию        
    *Сайт откуда собрана вакансия
По своему желанию можно добавить еще работодателя и расположение. Данная структура должна быть одинаковая для вакансий с обоих сайтов. Общий результат можно вывести с помощью dataFrame через pandas.

```python
import numpy as np
import pandas as pd
from pprint import pprint
import requests
import re
from bs4 import BeautifulSoup as bs
```

```python
vacancies_df = pd.read_csv('D:/AI/vacancies_superjob.csv')
```

```python
vacancies_df_hh = pd.read_csv('D:/AI/vacancies_hh.csv')
```

```python
sal_min = vacancies_df['Заработная плата'].str.findall(r'(\d+).(\d+).+')
```

```python
vacancies_df['min плата'] = sal_min
```

```python
#finding max salary from rows with min and max
sal_max = vacancies_df['Заработная плата'].str.findall(r'\d+.\d+.{3}(\d+).(\d+)')
```

```python

vacancies_df['max плата'] = sal_max
```

```python
#finding fixed salary (without range)
m = vacancies_df['Заработная плата'].str.findall(r'^(\d+).(\d+).₽')
```

```python
# adding fixed salary to MAXSALARY column
for i in range(len(vacancies_df['max плата'])):
    if vacancies_df['max плата'][i]:
        continue
    else:
        vacancies_df['max плата'][i] = m[i]
```

```python
for i in range(len(vacancies_df['max плата'])):
    if not vacancies_df['max плата'][i]:
        vacancies_df['max плата'][i] = np.nan
        continue
    else:
        vacancies_df['max плата'][i] = ''.join(vacancies_df['max плата'][i][0])
```

```python
for i in range(len(vacancies_df['min плата'])):
    if not vacancies_df['min плата'][i]:
        vacancies_df['min плата'][i] = np.nan
        continue
    else:
        vacancies_df['min плата'][i] = ''.join(vacancies_df['min плата'][i][0])
    
```

```python
pd.to_numeric(vacancies_df['max плата'])
pd.to_numeric(vacancies_df['min плата']);
```

```python
#find salary in USD
salary_usd = vacancies_df_hh['Заработная плата'].str.findall(r'.?(\d+.\d+)?-?(\d+.\d+).USD')
```

```python
# USD to rubles
for i in range(len(salary_usd)):

    if not salary_usd[i]:
        continue
    else:
        salary_usd[i] = list(salary_usd[i][0])
        salary_usd[i][0] = (salary_usd[i][0]).replace('\xa0', '')
        salary_usd[i][1] = (salary_usd[i][1]).replace('\xa0', '')

        salary_usd[i][0] = str((int(salary_usd[i][0]))*63)
        salary_usd[i][1] = str((int(salary_usd[i][1]))*63)
        
        print(f"USD salary in rubles{salary_usd[i]}")
```

```python
# adding MINSalary column to HH df
sal_min_hh = vacancies_df_hh['Заработная плата'].str.findall(r'от.(\d+).(\d+).+')
vacancies_df_hh['min плата'] = sal_min_hh
```

```python
for i in range(len(vacancies_df_hh['min плата'])):
    if not vacancies_df_hh['min плата'][i]:
        vacancies_df_hh['min плата'][i] = np.nan
        continue
    else:
        vacancies_df_hh['min плата'][i] = ''.join(vacancies_df_hh['min плата'][i][0])
    
```

```python
# adding MAXSalary column to HH df
sal_max_hh = vacancies_df_hh['Заработная плата'].str.findall(r'до.(\d+).(\d+).+')
vacancies_df_hh['max плата'] = sal_max_hh
```

```python
for i in range(len(vacancies_df_hh['max плата'])):
    if not vacancies_df_hh['max плата'][i]:
        vacancies_df_hh['max плата'][i] = np.nan
        continue
    else:
        vacancies_df_hh['max плата'][i] = ''.join(vacancies_df_hh['max плата'][i][0])
```

```python
# gathering salary with min and max (a range)
m_hh = vacancies_df_hh['Заработная плата'].str.findall(r'(\d+.\d+)-(\d+.\d+).руб.')

for i in range(len(m_hh)):
 
    if not m_hh[i]:
        continue
        
    else:
        m_hh[i] = list(m_hh[i][0])
        m_hh[i][0] = (m_hh[i][0]).replace('\xa0', '')
        m_hh[i][1] = (m_hh[i][1]).replace('\xa0', '')
```

```python
# adding max salary from salary range and converted USD salary to MAXSALARY column
for i in range(len(m_hh)):
    if m_hh[i]:
        vacancies_df_hh['max плата'][i] = m_hh[i][1]
    elif salary_usd[i]:
        vacancies_df_hh['max плата'][i] = salary_usd[i][1]
    else:
        continue
```

```python
# adding min salary from salary range and converted USD salary to MINSALARY column
for i in range(len(vacancies_df_hh['min плата'])):
    if m_hh[i]:
        vacancies_df_hh['min плата'][i] = m_hh[i][0]
    elif salary_usd[i]:
        vacancies_df_hh['min плата'][i] = salary_usd[i][0]
    else:
        continue
```
